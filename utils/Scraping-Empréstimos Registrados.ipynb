```python
# https://www.ivalor.com.br/empresas/acoes/emprestimo
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.b3.com.br/pt_br/produtos-e-servicos/emprestimo-de-ativos/renda-variavel/s_empreg_ativos/renda-variavel-8AA8D0CD701B61040170400B64840F04.htm?data=04/06/2020&f=0'
req = requests.get(url)
if req.status_code == 200:
    print('Requisição bem sucedida!')
    content = req.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find(name='table')
```

```python
table_str = str(table)
df = pd.read_html(table_str)[0]
```
